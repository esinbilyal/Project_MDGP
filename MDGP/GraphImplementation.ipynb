#### Read and store the data 

```python
# Read and store the data
import pandas as pd
correl_log_data = pd.read_csv('mtx_correl_log_ret.csv', header=0, index_col=0)

correl_log_data
```

```python
import pandas as pd
correl_ewm_vol_data = pd.read_csv('mtx_correl_ewm_vol.csv', header=0, index_col=0)

correl_ewm_vol_data
```

#### Implement the structure of the graphs

```python
import networkx as nx

G_log_ret = nx.Graph()

for i in correl_log_data.index:
    for j in correl_log_data.columns:
        if i != j and abs(correl_log_data.at[i, j]) > 0:
            G_log_ret.add_edge(i, j)

nx.draw(G_log_ret, with_labels=True)
```

```python
import networkx as nx

G_ewm_vol = nx.Graph()

for i in correl_ewm_vol_data.index:
    for j in correl_ewm_vol_data.columns:
        if i != j and abs(correl_ewm_vol_data.at[i, j]) > 0:
            G_ewm_vol.add_edge(i, j)

nx.draw(G_ewm_vol, with_labels=True)
```

#### Implement the rescaling of edges with weak correlation

```python
correl_log_data = abs(correl_log_data)
```

```python
correl_ewm_vol_data = abs(correl_ewm_vol_data)
```

```python
import numpy as np

# Initialize the white-list for log-return correlations
white_list_log = {}
for i in correl_log_data.index:
    row = correl_log_data.loc[i]
    indices = np.argpartition(row.values, -3)[-3:]
    for j in indices:
        if i != correl_log_data.columns[j]:
            if i in white_list_log:
                white_list_log[i].append(correl_log_data.columns[j])
            else:
                white_list_log[i] = [correl_log_data.columns[j]]

# Initialize the white-list for volatility correlations
white_list_vol = {}
for i in correl_ewm_vol_data.index:
    row = correl_ewm_vol_data.loc[i]
    indices = np.argpartition(row.values, -3)[-3:]
    for j in indices:
        if i != correl_ewm_vol_data.columns[j]:
            if i in white_list_vol:
                white_list_vol[i].append(correl_ewm_vol_data.columns[j])
            else:
                white_list_vol[i] = [correl_ewm_vol_data.columns[j]]
```

```python
# Rescale the graph for log-return correlations
G_log_ret = nx.Graph()
for i in correl_log_data.index:
    for j in correl_log_data.columns:
        if i != j and i in white_list_log and j in white_list_log[i]:
            G_log_ret.add_edge(i, j)

# Rescale the graph for volatility correlations
G_ewm_vol = nx.Graph()
for i in correl_ewm_vol_data.index:
    for j in correl_ewm_vol_data.columns:
        if i != j and i in white_list_vol and j in white_list_vol[i]:
            G_ewm_vol.add_edge(i, j)
```

In the code above, the np.argpartition function is used to quickly find the indices of the 3 largest values in each row of the DataFrames correl_log_data and `correl_vol

```python
nx.write_gexf(G_log_ret, 'rescaled_log_ret.gexf')
nx.write_gexf(G_ewm_vol, 'rescaled_ewm_vol.gexf')
```

```python
def kruskal_max_spanning_tree(graph):
    edges = list(graph.edges(data=True))
    edges.sort(key=lambda x: x[2].get('weight', 1), reverse=True)
    
    parent = {node: node for node in graph.nodes}
    rank = {node: 0 for node in graph.nodes}

    def find(node):
        if parent[node] != node:
            parent[node] = find(parent[node])
        return parent[node]
    
    def union(node1, node2):
        root1 = find(node1)
        root2 = find(node2)
        if root1 != root2:
            if rank[root1] > rank[root2]:
                parent[root2] = root1
            else:
                parent[root1] = root2
                if rank[root1] == rank[root2]:
                    rank[root2] += 1
    
    mst = []
    for edge in edges:
        node1, node2, weight = edge
        if find(node1) != find(node2):
            union(node1, node2)
            mst.append((node1, node2, weight))
    
    return mst

G_log_ret_mst = nx.Graph()
G_log_ret_mst.add_edges_from(kruskal_max_spanning_tree(G_log_ret))

G_ewm_vol_mst = nx.Graph()
G_ewm_vol_mst.add_edges_from(kruskal_max_spanning_tree(G_ewm_vol))
```
